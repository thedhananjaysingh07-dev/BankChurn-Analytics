```python
# Outlier detection using IQR
numerical_cols = ['CreditScore', 'Age', 'Tenure', 'Balance', 
                  'NumOfProducts', 'EstimatedSalary']

print("OUTLIER DETECTION (IQR Method)")
print("=" * 45)

for col in numerical_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    outliers = df[(df[col] < lower) | (df[col] > upper)]
    print(f"{col:<20} : {len(outliers)} outliers "
          f"(lower={lower:.1f}, upper={upper:.1f})")
```

```python
# Outlier Decision Documentation
print("OUTLIER TREATMENT DECISION")
print("=" * 45)
print("""
Age (359 outliers >62 yrs) : KEEP
  → Real customers, not data errors
  → Older segment is analytically valuable

NumOfProducts (60 with 4)  : KEEP  
  → High-product customers = bank's best segment
  → Removing them would hide important patterns

CreditScore (15 near 350)  : KEEP
  → Genuinely low scores, not entry errors
  → Valid range is 350-850, all within bounds

CONCLUSION: No outlier removal required.
Dataset remains at 10,000 rows.
""")
```

# Day 4 — Data Cleaning & Feature Engineering

```python
# Create a clean copy — never modify original dataframe
df_clean = df.copy()

# Drop non-analytical columns
df_clean = df_clean.drop(columns=['Year', 'CustomerId', 'Surname'])

print("Columns after dropping non-analytical fields:")
print(df_clean.columns.tolist())
print("\nShape:", df_clean.shape)
```

```python
# AGE SEGMENTS
df_clean['AgeGroup'] = pd.cut(df_clean['Age'],
                               bins=[0, 30, 45, 60, 100],
                               labels=['<30', '30-45', '46-60', '60+'])

# CREDIT SCORE BANDS
df_clean['CreditBand'] = pd.cut(df_clean['CreditScore'],
                                 bins=[0, 580, 720, 850],
                                 labels=['Low', 'Medium', 'High'])

# TENURE GROUPS
df_clean['TenureGroup'] = pd.cut(df_clean['Tenure'],
                                  bins=[-1, 2, 6, 10],
                                  labels=['New', 'Mid', 'Long'])

print("New segmentation columns added:")
print(df_clean[['Age','AgeGroup','CreditScore','CreditBand',
                'Tenure','TenureGroup']].head(10))
```

```python
# BALANCE SEGMENTS
df_clean['BalanceSegment'] = pd.cut(df_clean['Balance'],
                                     bins=[-1, 0, 50000, 125000, 300000],
                                     labels=['Zero', 'Low', 'Medium', 'High'])

# CLV SCORE (Customer Lifetime Value)
df_clean['CLV_Score'] = (df_clean['Balance'] * 0.5 + 
                          df_clean['EstimatedSalary'] * 0.3 + 
                          df_clean['Tenure'] * 1000 * 0.2)

# CLV SEGMENTS
df_clean['CLV_Segment'] = pd.qcut(df_clean['CLV_Score'],
                                    q=4,
                                    labels=['Low', 'Medium', 'High', 'Premium'])

# ENGAGEMENT SCORE
df_clean['EngagementScore'] = (df_clean['IsActiveMember'] * 
                                df_clean['NumOfProducts'])

print("Balance, CLV and Engagement columns added:")
print(df_clean[['Balance','BalanceSegment',
                'CLV_Score','CLV_Segment',
                'EngagementScore']].head())
```

```python
# BALANCE TO SALARY RATIO
df_clean['Balance_Salary_Ratio'] = (df_clean['Balance'] / 
                                     df_clean['EstimatedSalary']).round(2)

# Final check
print("Final cleaned dataset shape:", df_clean.shape)
print("\nAll columns:")
print(df_clean.columns.tolist())
print("\nSample of new columns:")
print(df_clean[['AgeGroup', 'CreditBand', 'TenureGroup', 
                 'BalanceSegment', 'CLV_Segment', 
                 'EngagementScore', 
                 'Balance_Salary_Ratio']].head())
```
